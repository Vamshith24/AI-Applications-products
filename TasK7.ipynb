```python
#A list in Python is an ordered, mutable collection of elements enclosed in square brackets [], where items are separated by commas.

#1. append() – Add an item to the end

fruits = ["apple", "banana", "mango"]

fruits.append("orange")

print(fruits)
     #Explanation

#append() adds a single element to the end of the list.
```

```python
#2. insert() – Add an item at a specific position

numbers = [10, 20, 40]

numbers.insert(2, 30)

print(numbers)
#Explanation

#insert(index, value) places the value at the specified index.
```

```python
#3. extend() – Add multiple elements

list1 = [1, 2, 3]
list2 = [4, 5, 6]

list1.extend(list2)

print(list1)

#Explanation

#extend() combines another iterable with the existing list.
```

```python
#4. remove() – Remove a specific element

colors = ["red", "blue", "green", "yellow"]

colors.remove("green")

print(colors)
#Explanation

#remove() deletes the first occurrence of the specified value.
```

```python
#5. pop() – Remove and return an element animals = ["cat", "dog", "rabbit"]

animals = ["cat", "dog", "rabbit"]

removed_item = animals.pop()

print("Removed:", removed_item)
print("Remaining list:", animals)

#Explanation

#pop() removes the last item by default and returns it.
```

```python
#6. clear() – Remove all elements

marks = [80, 90, 70, 85]

marks.clear()

print(marks)

#Explanation

#clear() empties the list.
```

```python
#7. sort() – Arrange elements in ascending order

scores = [45, 12, 78, 30]

scores.sort()

print(scores)

#Explanation

#sort() rearranges the elements in ascending order.
```

```python
#8. reverse() – Reverse the order of elements

letters = ['A', 'B', 'C', 'D']

letters.reverse()

print(letters)

#Explanation

#reverse() changes the order of the list from last to first.
```

```python
#9. count() – Count occurrences of an element

numbers = [1, 2, 3, 2, 4, 2, 5]

count_of_two = numbers.count(2)

print("Number of times 2 appears:", count_of_two)

#Explanation

#count(value) returns how many times the value occurs in the list.
```

```python
#10. index() – Find the position of an element

cities = ["Delhi", "Mumbai", "Chennai", "Bangalore"]

position = cities.index("Chennai")

print("Index of Chennai:", position)

#Explanation

#index(value) returns the position of the first occurrence of the value.
```

```python
#  Unlike lists, tuples are immutable, so they have only two built-in methods:

   #count()
   #index()

#1. count() – Count occurrences of a value

numbers = (1, 2, 3, 2, 4, 2)

result = numbers.count(2)

print("Count of 2:", result)
```

```python
#2. count() with strings

fruits = ("apple", "banana", "apple", "orange")

result = fruits.count("apple")

print("Number of apples:", result)
```

```python
#3. count() when element is absent

colors = ("red", "blue", "green")

result = colors.count("yellow")

print(result)
```

```python
#4. index() – Find position of an element

animals = ("cat", "dog", "rabbit", "horse")

position = animals.index("rabbit")

print("Index:", position)
```

```python
#5. Tuple slicing

marks = (85, 90, 78, 92, 88)

new_tuple = marks[1:4]

print(new_tuple)
```

```python
#6. Tuple concatenation

tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)

result = tuple1 + tuple2

print(result)
```

```python
#7. Tuple repetition

letters = ("A", "B")

result = letters * 3

print(result)
```

```python
#8. Tuple unpacking

student = ("Rahul", 21, "Computer Science")

name, age, course = student

print("Name:", name)
print("Age:", age)
print("Course:", course)


#Explanation

#Tuple unpacking assigns tuple values to multiple variables.
```

```python
#In Python, a dictionary is a built-in data type used to store data in key–value pairs.

#Think of it like a real dictionary:

# a word = key
# its meaning = value

#1. get() Method

#Returns the value for a key. If the key doesn't exist, it returns a default value

student = {
    "name": "Alice",
    "age": 20
}

print(student.get("name"))
print(student.get("grade", "Not Available"))
```

```python
#2. keys() Method

#Returns all the keys in the dictionary.

student = {
    "name": "Alice",
    "age": 20,
    "city": "New York"
}

print(student.keys())
```

```python
#3. values() Method

#Returns all the values in the dictionary.

student = {
    "name": "Alice",
    "age": 20,
    "city": "New York"
}

print(student.values())
```

```python
#4. items() Method

#Returns key-value pairs as tuples.

student = {
    "name": "Alice",
    "age": 20,
    "city": "New York"
}

print(student.items())
```

```python
#5. update() Method

#Adds new items or updates existing ones.

student = {
    "name": "Alice",
    "age": 20
}

student.update({"city": "New York"})
student.update({"age": 21})

print(student)
```

```python
#6. pop() Method

#Removes a specified key and returns its value.

student = {
    "name": "Alice",
    "age": 20,
    "city": "New York"
}

removed_value = student.pop("city")

print("Removed:", removed_value)
print(student)
```

```python
#7. setdefault() Method

#Returns the value of a key. If the key doesn't exist, it inserts the key with the given value.

student = {
    "name": "Alice",
    "age": 20
}

student.setdefault("grade", "A")

print(student)
```

```python
#8. copy() Method

#Creates a copy of the dictionary.

student = {
    "name": "Alice",
    "age": 20
}

new_student = student.copy()

print("Original:", student)
print("Copy:", new_student)
```

```python
#9. clear() Method

#Removes all items from the dictionary.

student = {
    "name": "Alice",
    "age": 20,
    "city": "New York"
}

student.clear()

print(student)
```
