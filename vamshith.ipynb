```python
# ==================================
# Topic: Introduction to Python
# ==================================

# Explanation:
# Python is a high-level interpreted programming language.

# Syntax:
# print("message")

# Example:
print("Hello Python")   # Displays message
```

```python
# ==================================
# Topic: Data Types
# ==================================

# Explanation:
# Data types define the type of value stored in a variable.

# Syntax:
# variable_name = value

# Example:

age = 20          # Integer value
name = "John"     # String value
price = 99.5      # Float value
result = True     # Boolean value

print(age)
print(name)
print(price)
print(result)
```

```python
# ==================================
# Topic: Integer (int)
# ==================================

# Explanation:
# Integer stores whole numbers.

# Syntax:
# variable = integer_value

# Example:

number = 100      # Creating integer variable

print(number)     # Printing value
```

```python
# ==================================
# Topic: Float
# ==================================

# Explanation:
# Float stores decimal numbers.

# Syntax:
# variable = decimal_value

# Example:

height = 5.8      # Float variable

print(height)     # Displaying float
```

```python
# ==================================
# Topic: Boolean
# ==================================

# Explanation:
# Boolean stores True or False values.

# Syntax:
# variable = True/False

# Example:

is_python = True  # Boolean value

print(is_python)
```

```python
# ==================================
# Topic: String
# ==================================

# Explanation:
# String stores text characters.

# Syntax:
# variable = "text"

# Example:

name = "vamshi"  # String variable

print(name)
```

```python
# ==================================
# Topic: List
# ==================================

# Explanation:
# List stores multiple values in one variable.

# Syntax:
# list_name = [value1,value2]

# Example:

marks = [80,90,75]   # Creating list

print(marks)

print(marks[0])      # Access first value
```

```python
# ==================================
# Topic: Variables
# ==================================

# Explanation:
# Variable stores data values.

# Syntax:
# variable_name = value

# Example:

x = 50       # Creating variable

print(x)

x = 100      # Updating value

print(x)
```

```python
# ==================================
# Topic: Expressions
# ==================================

# Explanation:
# Expression produces a result using operators.

# Syntax:
# value operator value

# Example:

total = 10 + 5    # Addition expression

print(total)
```

```python
# ==================================
# Topic: Statements
# ==================================

# Explanation:
# Statement is a command executed by Python.

# Syntax:
# statement

# Example:

print("Hello")    # Print statement

x = 20            # Assignment statement

print(x)
```

```python
# ==================================
# Topic: Precedence of Operators
# ==================================

# Explanation:
# Precedence decides operator execution order.

# Syntax:
# ()
# **
# *
# /
# +
# -

# Example:

result = 10 + 5 * 2

print(result)     # Multiplication first
```

```python
# ==================================
# Topic: Comments
# ==================================

# Explanation:
# Comments are ignored by Python interpreter.

# Syntax:
# # comment

# Example:

# This is a comment

print("Python")
```

```python
# ==================================
# Topic: Conditional (if)
# ==================================

# Explanation:
# if statement executes code when condition is True.

# Syntax:
# if condition:
#     statements

# Example:

age = 18

if age >= 18:
    # Executes when condition is true
    print("Eligible to vote")
```

```python
# ==================================
# Topic: Alternative (if-else)
# ==================================

# Explanation:
# if-else selects one block based on condition.

# Syntax:
# if condition:
#     statements
# else:
#     statements

# Example:

age = 15

if age >= 18:
    print("Adult")
else:
    print("Not Adult")
```

```python
# ==================================
# Topic: Chained Conditional (if-elif-else)
# ==================================

# Explanation:
# elif checks multiple conditions.

# Syntax:
# if condition:
# elif condition:
# else:

# Example:

marks = 75

if marks >= 90:
    print("A Grade")

elif marks >= 60:
    print("B Grade")

else:
    print("Fail")
```

```python
# ==================================
# Topic: Boolean Values and Operators
# ==================================

# Explanation:
# Boolean operators combine multiple conditions.

# Syntax:
# condition1 and condition2
# condition1 or condition2
# not condition

# Example:

age = 20
citizen = True

if age >= 18 and citizen:
    print("Can vote")
```

```python
# ==================================
# Topic: While Loop
# ==================================

# Explanation:
# while loop repeats code until condition becomes false.

# Syntax:
# while condition:
#     statements

# Example:

i = 1

while i <= 5:
    print(i)

    # Increasing value
    i = i + 1
```

```python
# ==================================
# Topic: For Loop
# ==================================

# Explanation:
# for loop repeats code for each item in sequence.

# Syntax:
# for variable in sequence:
#     statements

# Example:

for i in range(5):

    # Printing numbers
    print(i)
```

```python
# ==================================
# Topic: Break Statement
# ==================================

# Explanation:
# break stops loop execution immediately.

# Syntax:
# break

# Example:

for i in range(10):

    if i == 5:
        break       # Stops loop

    print(i)
```

```python
# ==================================
# Topic: Continue Statement
# ==================================

# Explanation:
# continue skips current loop iteration.

# Syntax:
# continue

# Example:

for i in range(5):

    if i == 2:
        continue    # Skip 2

    print(i)
```

```python
# ==================================
# Topic: Fruitful Functions (Return Values)
# ==================================

# Explanation:
# Function that returns a value is called fruitful function.

# Syntax:
# def function():
#     return value

# Example:

def add():

    # Returning value
    return 10 + 20

result = add()

print(result)
```

```python
# ==================================
# Topic: Function Parameters
# ==================================

# Explanation:
# Parameters allow functions to receive values.

# Syntax:
# def function(parameter):

# Example:

def square(number):

    return number * number

print(square(5))
```

```python
# ==================================
# Topic: Local and Global Scope
# ==================================

# Explanation:
# Scope defines where a variable can be accessed.

# Example:

x = 100     # Global variable

def test():

    y = 50  # Local variable

    print(y)

test()

print(x)
```

```python
# ==================================
# Topic: Function Composition
# ==================================

# Explanation:
# Using one function result inside another function.

# Syntax:
# function1(function2())

# Example:

def add(x):
    return x + 5

def multiply(y):
    return y * 2

print(multiply(add(10)))
```

```python
# ==================================
# Topic: Recursion
# ==================================

# Explanation:
# Function calling itself is called recursion.

# Syntax:
# def function():
#     function()

# Example:

def count(n):

    if n == 0:
        return

    print(n)

    count(n-1)

count(5)
```
